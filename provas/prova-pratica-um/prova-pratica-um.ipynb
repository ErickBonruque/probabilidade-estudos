```python
import pandas as pd
```

```python
dados = pd.read_excel('ethereum.xlsx')
```

```python
dados.info()
```

```python
dados.head()
```

```python
dados = pd.DataFrame(dados)
dados
```

```python
amostraSimples = dados.sample(n=36)
amostraSimples
```

```python
volume = []
for i in amostraSimples['volume']:
    i = i.replace('K', '')  # remove o 'K'
    i = i.replace(',', '.')  # substitui a vírgula por um ponto
    i = i.replace('M', '')  # remove o 'M'
    i = i.replace(' ', '')  # remove o ' '
    i = float(i)  # converte para float
    volume.append(i)
```

```python
intervalo_volume = pd.cut(volume, bins=5)
print(intervalo_volume)
```

```python
frequncia_absoluta = pd.value_counts(intervalo_volume)
frequncia_absoluta
```

```python
tabela = pd.DataFrame({'Intervalo de Volume': frequncia_absoluta.index, 'Frequência Absoluta': frequncia_absoluta.values})
tabela
```

## Estatistica Descritiva

O coeficiente de variação em 54% mostra que a uma grande variação nos dados em relação a media geral dos valores

```python
volume = pd.Series(volume)
print("Media = ", volume.mean().round(2))
print("Mediana = ", volume.median().round(2))
print("Moda =", volume.mode()[0].round(2))
print("Variancia = ", volume.var().round(2))
print("Desvio Padrão = ", volume.std().round(2))
print("Coeficiente de Variação = ", (volume.std()/volume.mean()).round(2))
```

```python
import matplotlib.pyplot as plt
plt.boxplot(volume)
plt.title('Boxplot do Volume')
plt.show()
```

```python
plt.hist(volume, bins=5)
plt.title('Histograma do Volume')
plt.show()
```

```python
from scipy.stats import skew, kurtosis

volume_skewness = skew(volume)
volume_kurtosis = kurtosis(volume)

print("Assimetria do Volume = ", volume_skewness.round(3))
print("Curtose do Volume = ", volume_kurtosis.round(3))
```
