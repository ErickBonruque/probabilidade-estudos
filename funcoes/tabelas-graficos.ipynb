# Contrução De Tabelas e Graficos

## Tabela

```python
import pandas as pd
```

```python
import random
escolaridade = ['Primario', 'Secundario', 'Superior', 'Pos-Graduacao', 'Mestrado', 'Doutorado']
total = [random.randint(0, 2000) for _ in range(len(escolaridade))] # gera uma lista com valores aleatórios
total
```

```python
tabela = pd.DataFrame({'Escolaridade': escolaridade, 'Total': total})
tabela
```

```python
tabela['Porcentagem'] = tabela['Total'] / tabela['Total'].sum() * 100
tabela['Porcentagem'] = tabela['Porcentagem'].round(2)
tabela
```

Certinho viado, mais alunos no Doutorado do que no primario

## Graficos

```python
import matplotlib.pyplot as plt
```

```python
plt.pie(tabela['Porcentagem'], labels=tabela['Escolaridade'], autopct='%1.1f%%')
plt.title('Escolaridade dos Funcionários')
plt.show()
```

```python
tabela.plot(x='Escolaridade', y='Total', kind='bar')
plt.xlabel('Escolaridade')
plt.ylabel('Total')
plt.title('Total por Escolaridade')
plt.show()
```
